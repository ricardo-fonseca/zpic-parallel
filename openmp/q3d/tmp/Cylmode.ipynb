# Cylindrical modes

## Real part of azimuthal modes

Blue lines represent positive values, orange lines represent negative values

```python
import numpy as np
import matplotlib.pyplot as plt


def plot_mode( m, ax ):
    npoints = 1024
    angle = np.linspace( 0, 2*np.pi, npoints, endpoint = True )

    r = np.cos(m * angle)

    pos  = np.ma.masked_where( r >= 0, angle )
    posr = -np.ma.masked_where( r >= 0, r )

    neg  = np.ma.masked_where( r < 0, angle )
    negr = np.ma.masked_where( r < 0, r )

    c = ax.plot( neg, negr )
    c = ax.plot( pos, posr )
    ax.set( title = 'm = {}'.format(m))


fig, axs = plt.subplots( 2, 2, 
        gridspec_kw = dict( hspace = 0.6, top = 1.1 ),
        subplot_kw = dict(projection='polar'))

plot_mode( 0, axs[0,0] )
plot_mode( 1, axs[0,1] )
plot_mode( 2, axs[1,0] )
plot_mode( 3, axs[1,1] )

# plt.savefig("cyl_modes.svg", format="svg", bbox_inches='tight', transparent=True)

plt.show()
```

## Imaginary part of azimuthal modes

Blue lines represent positive values, orange lines represent negative values

```python
import numpy as np
import matplotlib.pyplot as plt


def plot_imode( m, ax ):
    npoints = 1024
    angle = np.linspace( 0, 2*np.pi, npoints, endpoint = True )

    r = np.sin(m * angle)

    pos  = np.ma.masked_where( r >= 0, angle )
    posr = -np.ma.masked_where( r >= 0, r )

    neg  = np.ma.masked_where( r < 0, angle )
    negr = np.ma.masked_where( r < 0, r )

    c = ax.plot( neg, negr )
    c = ax.plot( pos, posr )
    ax.set_ylim([0,1.1])
    ax.set( title = 'm = {}'.format(m))


fig, axs = plt.subplots( 2, 2, 
        gridspec_kw = dict( hspace = 0.6, top = 1.1 ),
        subplot_kw = dict(projection='polar'))

plot_imode( 0, axs[0,0] )
plot_imode( 1, axs[0,1] )
plot_imode( 2, axs[1,0] )
plot_imode( 3, axs[1,1] )

# plt.savefig("cyl_modes.svg", format="svg", bbox_inches='tight', transparent=True)

plt.show()
```
