## Grid test

`test_grid()`

This test implements a simple Poisson solver

```python
import visxd
import zdf
import matplotlib.colors as colors

visxd.grid2d( "charge.zdf")

info = zdf.info( "charge_k.zdf" )

# Shift ky = 0 to vertical center
shifty = int( info.grid.nx[1] // 2 ) - 1

visxd.complex_grid2d( "charge_k.zdf", part = 'abs', shift = [0,shifty], norm = colors.LogNorm())
visxd.complex_grid2d( "potential_k.zdf", part = 'abs', shift = [0,shifty], norm = colors.LogNorm() )

visxd.grid2d( "potential.zdf" )
```
