## ZPIC paper

Weibel plots

```python
import visxd
import matplotlib.pyplot as plt
import matplotlib.colors as colors

plt.rcParams.update({'font.size': 16})

map = "viridis_r"
#map = "YlOrBr"
#map = "cividis_r"

# visxd.plot_data("CHARGE/positrons/positrons-charge", 50 , vmin = 0, vmax = 9, cmap = map, save = "t035.pdf" )
# visxd.plot_data("CHARGE/positrons/positrons-charge", 260, vmin = 0, vmax = 9, cmap = map, save = "t182.pdf" )
# visxd.plot_data("CHARGE/positrons/positrons-charge", 500, vmin = 0, vmax = 9, cmap = map, save = "t350.pdf" )

def over_grid( data, iter, save = False ):
    visxd.plot_data(data, iter , vmin = 0, vmax = 9, cmap = map, show = False )
    for i in range(1,8):
        plt.axhline(y = i * 32 * 0.1, linewidth = 0.5, color = 'k', linestyle="--")
        plt.axvline(x = i * 32 * 0.1, linewidth = 0.5, color = 'k', linestyle="--")
    if ( save ):
        plt.savefig(save, format="pdf", bbox_inches="tight")
    plt.show()

over_grid( "CHARGE/positrons/positrons-charge", 50 , save = "t035.pdf" )
over_grid( "CHARGE/positrons/positrons-charge", 260, save = "t182.pdf" )
over_grid( "CHARGE/positrons/positrons-charge", 500, save = "t350.pdf" )
```
